**Warp Perspective / Bird View**

In warp perspective we will convert an image from a perspective to a bird view.Using **OpenCV method**



```python
import cv2
import numpy as np
```

```python
img = cv2.imread('./perspectiveImg.jpg')
pts1 = np.float32([[132,331],[736,264],[308, 979],[944,756]])
print(pts1)
```

```python
width, height = 250,350
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])
matrix = cv2.getPerspectiveTransform(pts1,pts2)
imgOutput = cv2.warpPerspective(img,matrix,(width,height))
```

```python
for x in range(0,4):
    cv2.circle(img,(int(pts1[x][0]),int(pts1[x][1])),5,(0,0,255),cv2.FILLED)
```

```python
cv2.imshow("Original image", img)
cv2.imshow("Output image", imgOutput)
cv2.waitKey(0)
```
